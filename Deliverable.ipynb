**Video:** https://www.tiktok.com/@cees_evryting/video/7569032934200511799?q=gym%20attire%20controversy&t=1762792093968

**Description:** The video is about a locker-room incident at a Gold’s Gym in the Beverly Center in Los Angeles. It says that on November 3, 2025, Tish Hyman confronted Grant Kyle Freeman in the women’s locker room, describing Freeman as “a man identifying as a woman,” and that Hyman felt violated while changing. The gym revoked Hyman’s membership and had her escorted out by police, while Freeman faced no repercussions under the gym’s gender-identity policy. The slide claims the incident sparked widespread online support for Hyman, including boycott calls and negative reviews aimed at the gym. It also notes that reactions have been polarized, especially in debates around trans rights.

**Research Question:** Does the audience of this video believe that biological males should have access to women's locker rooms?

**Concept 1:** Yes, biological males should have access to women's locker rooms. 

**Concept 2:** No, biological males should not have access to women's locker rooms.

**First LLM Prompt:** You are a research assistant. 
                You will receive a series of tiktok comments, 
                please code for if biological males should have access to womens locker rooms or not. 
                
**Inter-rater reliability score:** concept 1: -0.297 concept 2: 0.342

**Final LLM Prompt:** You are a research assistant. 
                You will receive a series of tiktok comments, 
                please code for if biological males should have access to womens locker rooms or not. Concept 1: Yes, biological men should have access to womens locker rooms code 1. Also code 1 if the words "unisex", "family bathroom", "gender neutral" come up in a comment, code 0 for all else. Concept 2: No, biological men should not have access to women's locker rooms code 1. If comments say "No" "absolutely not" "unisex" "gender neutral" or "family bathroom" code 1. Code 0 for all else 

**Inter-rater reliability score:** concept 1: 0.730 concept 2: 0.703

## Final Write Up:


Our project looked at how people reacted to a TikTok about a locker-room incident at a Gold’s Gym in Los Angeles, where Tish Hyman confronted Grant Kyle Freeman in the women’s locker room and was later banned from the gym. Our main question was: **Does the audience of this video believe that biological males should have access to women’s locker rooms?** To study this, we coded the comments using two concepts: Concept 1 was “Yes, biological males should have access to women’s locker rooms”, and Concept 2 was “No, biological males should not have access to women’s locker rooms” (. With our first, more vague LLM prompt, the model didn’t line up very well with our human coding, and the inter-rater reliability scores were low (concept 1: -0.297; concept 2: 0.342).

We then updated the prompt to be way more specific about what should count as a 1 for each concept, especially around certain key phrases like “unisex,” “gender neutral,” and “family bathroom.” After that change, the agreement between the LLM and human coders went up a lot (concept 1: 0.730; concept 2: 0.703), which basically confirmed what we talked about in lecture: if we want LLMs to act like decent research assistants, we have to spell out our coding rules very clearly. Even so, we noticed that we still needed humans to actually watch the video and interpret vague “yes” or “no” comments, because the LLM doesn’t know what those replies are referring to without that extra context. So, the model got more consistent, but it still couldn’t fully replace human judgment.

In terms of what the audience actually thinks, our results suggest that **most commenters do not support** allowing biological males to use women’s locker rooms. Concept 2 (opposition) got the most 1’s overall, while concept 1 (support + unisex/gender-neutral suggestions) showed up less often. However, comments aligned with concept 1 tended to get **more replies**, which we think is because that position is seen as more controversial and therefore draws more pushback and debate. When we looked at age patterns, commenters who didn’t clearly align with either concept were mostly in the 25–34 range. Those who aligned with concept 1 were generally on the younger side of 18–24 (likely 18–21), while those aligned with concept 2 were also 18–24 but slightly older on average (around 21–24). Overall, we conclude that this TikTok audience leans against allowing biological males in women’s locker rooms, but there’s a smaller, very vocal group arguing for access or for unisex/gender-neutral alternatives.
